```python
#using function return true if 2 list have atleast one common value
def common_member(list1, list2):
    for item in list1:
        if item in list2:
            return True
    return False

list1 = [1, 2, 3, 4]
list2 = [5, 6, 3, 8]

print(common_member(list1, list2))
```
